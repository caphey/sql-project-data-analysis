## Import des librairies

```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv
import os
```

## Création de la connexion à la base de données PostgreSQL

```python
def get_connection():
    return psycopg2.connect(
        dbname=os.getenv('DB_NAME'),
        user=os.getenv('DB_USER'),
        password=os.getenv('DB_PASSWORD'),
        host=os.getenv('DB_HOST')
    )
```

Création d'une fonction pour lire chaque requête SQL provenant du fichier `sql_queries.py`

```python
def fetch_data(query):
    connection = get_connection()
    df = pd.read_sql_query(query, connection)
    connection.close()
    return df
```

## Exploration des données

```python
table_names = [
    'film',
    'serie',
    'evaluation',
    'abonnement',
    'genre',
    'sortie',
    'utilisateur',
    'episode_serie',
    'realisateur'
]
```

```python
for table_name in table_names:
    try:
        query = f"SELECT * FROM {table_name}"
        df = fetch_data(query)
        print(f"Table: {table_name}")
        print(df.info())
        print()
    except Exception as e:
        print(f"Error while fetching data from table {table_name}")
        print(e)
```

## Visualisation de données

```python
from sql_queries import GET_MOVIES_BY_GENRE

df = fetch_data(GET_MOVIES_BY_GENRE)

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='genre', y='nombre_films', palette='viridis')
plt.title('Nombre de films par genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de films')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sql_queries import GET_ALL_MOVIES_BY_DATE

df = fetch_data(GET_ALL_MOVIES_BY_DATE)

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='num_films', marker='o' )
plt.title('Nombre de films sortis avant une date donnée')
plt.xlabel('Date de sortie')
plt.ylabel('Nombre de films')
plt.title('Nombre de films sorties par année')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sql_queries import GET_DURATION_SERIES

df = fetch_data(GET_DURATION_SERIES)

plt.figure(figsize=(12, 6))
sns.boxplot(x='genre', y='duree_serie', data=df, palette='viridis')
plt.xlabel('Genre')
plt.ylabel('Durée des séries (en minutes)')
plt.title('Durée des séries par genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


Les genres "Action" et "Thriller" ont une durée de série assez similaire, légèrement inférieure à la moyenne.
Les genres "Animation" et "Documentaire" ont une durée de série plus courte par rapport aux autres genres.
Il est intéressant de noter que les genres "Romance" et "Biographie" ont une plus grande variabilité dans la durée des séries, avec des boîtes à moustaches plus étendues, indiquant une plus grande dispersion des valeurs.

En général, ce graphique permet de visualiser les différences de durée des séries entre les différents genres, ce qui peut être utile pour les analyses et les recommandations basées sur la durée des séries.

```python
from sql_queries import GET_SERIES

df = fetch_data(GET_SERIES)

plt.figure(figsize=(12, 6))
plt.scatter(df['nombre_episodes'], df['duree'], alpha=0.5, s=100, color='green')
plt.xlabel('Nombre d\'épisodes')
plt.ylabel('Durée de la série')
plt.title('Nombre d\'épisodes et durée des séries')
plt.tight_layout()
plt.show()
```

Ce graphique montre la relation entre le nombre d'épisodes et la durée des séries. On peut voir que la plupart des séries ont moins de 100 épisodes et durent environ 50 minutes.

```python
from sql_queries import GET_REALISATIONS_BY_REALISATOR

df = fetch_data(GET_REALISATIONS_BY_REALISATOR)

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='realisateur', y='nb_real', palette='viridis')
plt.xlabel('Réalisateur')
plt.ylabel('Nombre de réalisations')
plt.title('Nombre de réalisations par réalisateur')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sql_queries import GET_MEAN_NOTE_BY_SERIE, GET_MEAN_NOTE_BY_GENRE, GET_MEAN_NOTE_BY_MOVIE

df_serie = fetch_data(GET_MEAN_NOTE_BY_SERIE)
df_genre = fetch_data(GET_MEAN_NOTE_BY_GENRE)
df_film = fetch_data(GET_MEAN_NOTE_BY_MOVIE)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

sns.barplot(data=df_serie, x='serie', y='note_moyenne', palette='viridis', ax=ax[0])
ax[0].set_xlabel('Series')
ax[0].set_ylabel('Note moyenne')
ax[0].set_title('Note moyenne des séries')
ax[0].tick_params(axis='x', rotation=45)

sns.barplot(data=df_film, x='film', y='note_moyenne', palette='viridis', ax=ax[1])
ax[1].set_xlabel('Genre')
ax[1].set_ylabel('Note moyenne')
ax[1].set_title('Note moyenne des genres')
ax[1].tick_params(axis='x', rotation=45)

sns.barplot(data=df_genre, x='genre', y='note_moyenne', palette='viridis', ax=ax[2])
ax[2].set_xlabel('Film')
ax[2].set_ylabel('Note moyenne')
ax[2].set_title('Note moyenne des films')
ax[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
```

Ces graphiques donnent un aperçu de la popularité et de la qualité des séries, des genres et des films sur la base des notes moyennes.

On peut observer que certaines séries, certains genres et certains films sont systématiquement mieux notés, ce qui témoigne de leur attrait pour le public.
Ces informations peuvent être utiles pour les recommandations

```python
from sql_queries import GET_MOVIES_SERIES_WATCHED_BY_USER

df = fetch_data(GET_MOVIES_SERIES_WATCHED_BY_USER)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='filmsregarde', y='seriesregarde', s=100, alpha=0.5, color='green')
plt.xlabel('Films regardés')
plt.ylabel('Séries regardées')
plt.title('Films et séries regardés par utilisateur')
plt.tight_layout()
plt.show()
```

```python
from sql_queries import GET_EVALUATION_BY_GENRE_AND_YEAR

df = fetch_data(GET_EVALUATION_BY_GENRE_AND_YEAR)

df = df.pivot(index='genre', columns='year', values='note_moyenne')

plt.figure(figsize=(12, 6))
sns.heatmap(data=df, cmap='viridis')
plt.title('Note moyenne par genre et par année')
plt.xlabel('Année')
plt.ylabel('Genre')
plt.tight_layout()
```

Ce graphique fournit une représentation visuelle de la manière dont les notes d'évaluation moyennes varient en fonction des genres et des années.
Il nous permet d'identifier des tendances et des modèles dans les notes d'évaluation au fil du temps et entre les genres :

Certains genres ont systématiquement des notes d'évaluation moyennes plus élevées sur plusieurs années, d'autres présentent des variations dans les scores d'évaluations moyens au fil des ans.
